```
#N numero de hombres, se asume el número de mujeres es igual
def wPrefersM1OverM(prefer, w, m, m1, N): #indica si una mujer prefiere a M1 sobre M2, hombres
	for i in range(N): 
		if (prefer[w][i] == m1): 
			return True
		if (prefer[w][i] == m): 
			return False

def stableMarriage(prefer, N): #Implementa el algoritmo Gale-shapley asumiendo que los hombres se proponen a las mujeres 
	wPartner = [-1 for i in range(N)] 
	mFree = [False for i in range(N)] 

	freeCount = N 
	while (freeCount > 0): 
		m = 0
		while (m < N): 
			if (mFree[m] == False): 
				break
			m += 1
		i = 0
		while i < N and mFree[m] == False: 
			w = prefer[m][i] 
			if (wPartner[w - N] == -1): 
				wPartner[w - N] = m 
				mFree[m] = True
				freeCount -= 1

			else: 
				m1 = wPartner[w - N] 
				if (wPrefersM1OverM(prefer, w, m, m1, N) == False): 
					wPartner[w - N] = m 
					mFree[m] = True
					mFree[m1] = False
			i += 1
	print("Mujer ", " Hombre") 
	for i in range(N): 
		print(i + N, "\t", wPartner[i]) 
```

#El mejor caso para el algoritmo Gale-shapley ocurre cuando un hombre se propone y nunca es rechazado. Lo anterior ocurre cuando cada hombre se propone a una mujer diferente. 
#Se sabe que el algoritmo termina cuando toda mujer ha recibido una propuesta, por lo tanto, en el peor de los casos podriamos pensar que existe una mujer que nunca recibe una propuesta si no hasta la última iteración. El resto de mujeres, en el peor de los casos, reciben tantas propuestas como hay hombres, es decir si el número de hombres es n, cada una de estas mujeres recibiría n propuestas. Así, el número de propuestas sería n*(n-1) + 1, n-1 mujeres reciben n propuestas, y un única mujer recibe una propuesta.

```
def get_ranking_matrix(preference): #Returns the rank matrix which is used during the selection
    rank = [[None for j in range(len(preference))] for i in range(len(preference))]
    for i in range(len(preference)):
        for j in range(len(preference[i])):
            rank[i][preference[i][j]] = j
    return rank

def stable_roommates_phase_1(preference, rank): #In the first phase, everybody proposes to its favorite. Each person should end with the proposition of someone, and having proposed to someone else who acepted
                                                #If the above condtion doesn't fulfill then the algorithm ends and it means there isn't a stable matching
    proposal = [None for x in range(len(preference))]
    first = [0 for x in range(len(preference))]
    last = [len(x) for x in preference]
    to_process = [x for x in range(len(preference))]
    while len(to_process) > 0:
        i = to_process[0]
        while preference[i][first[i]] == None:
            first[i] += 1
            
        top_pick = preference[i][first[i]]
        if proposal[top_pick] == None:
            proposal[top_pick] = i
            
            match_rank = preference[top_pick].index(i)
            for x in range(match_rank+1, last[top_pick]):
                reject = preference[top_pick][x]
                preference[reject][(rank[reject][top_pick])] = None
            last[top_pick] = match_rank
            del to_process[0]
            
            continue
        
        curr_match_idx = rank[top_pick][proposal[top_pick]]
        potential_match_idx = rank[top_pick][i]
        
        if curr_match_idx < potential_match_idx: # current matching is preferred, i is rejected
            preference[top_pick][potential_match_idx] = None
            
            first[i] += 1 # start at next spot
            
            continue
        else: # accept proposal, so old match has to return to their preference list again
            preference[top_pick][curr_match_idx] = None
            
            # old match is rejected by top_pick, must update their list
            top_pick_idx = rank[proposal[top_pick]][top_pick]
            preference[proposal[top_pick]][top_pick_idx] = None
            
            del to_process[0]
            # add old match to to_process
            to_process.insert(0, proposal[top_pick])
            
            proposal[top_pick] = i
            last[top_pick] = potential_match_idx
    
    return first, last, preference

def clean_preferences(first, last, preferences): #In the second phase, everyone rejects all potential partnerts less desirable than their current
                                                  #accepted one, this is done symetrically. 
    for i in range(len(preferences)):
        for j in range(len(preferences[i])):
            if j < first[i] or j > last[i]:
                preferences[i][j] = None

    return preferences

def find_second_favorite(i, first, last, pref):
    count = 0
    for j in range(first[i], last[i]+1):
        if not pref[j] == None:
            count += 1
        elif count == 0:
            first[i] += 1
        if count == 2:
            return pref[j]
    return None

def find_rotation(i, p, q, first, last, preferences): #In the third phase, we find each participant which has more than one choice, using their second preference
                                                      #we try to find a cycle which means that the above choices must be rejected, this procedure ends when everyone has one choice, and that's 
                                                      #the stable matching found. 
    second_favorite = find_second_favorite(p[i], first, last, preferences[p[i]])
    next_p = preferences[second_favorite][last[second_favorite]]
    
    if next_p in p:
        # rotation found!
        j = p.index(next_p)
        q[j] = second_favorite
                
        return p[j:], q[j:]

    q.append(second_favorite)
    p.append(next_p)
    return find_rotation(i+1, p, q, first, last,  preferences)

def eliminate_rotation(p, q, first, last, preferences, rank):
    for i in range(len(p)):
        # q_i rejects p_i so that p_i proposes to q_i+1
        preferences[p[i]][rank[p[i]][q[i]]] = None
        
        # all successors of p_i-1 are removed from q_i's list, and q_i is removed from their lists
        for j in range(rank[q[i]][p[i-1]]+1, last[q[i]]):
            reject = rank[q[i]].index(j) #preferences[q[i]][j]
            preferences[reject][rank[reject][q[i]]] = None
            
        last[q[i]] = rank[q[i]][p[i-1]]

def stable_roommates_phase_2(first, last, preferences, rank):
    while True:
        p, q = None, None
        # find first p_0 to get a rotation from
        # preference list of p_0 must contain at least 2 elements
        for i in range(len(preferences)):
            if last[i] - first[i] > 0 and find_second_favorite(i, first, last, preferences[i]) != None:
                p, q = find_rotation(0, [i], [None], first, last, preferences)
                break
        
        if not p and not q:
            return preferences
        
        # eliminate rotation
        eliminate_rotation(p, q, first, last, preferences, rank)

def match_roommates(preferences):
    rank = get_ranking_matrix(preferences)
    first, last, preferences = stable_roommates_phase_1(preferences, rank)
    stable_roommates_phase_2(first, last, preferences, rank)
    clean_preferences(first, last, preferences)
    
    matches = []
    length = len(preferences)
    visited = set()
    i = 0
    
    for i in range(len(preferences)):
        if not i in visited:
            pair = (i, preferences[i][last[i]])
            visited.add(last[i])
            matches.append(pair)
    
    return matches
```

```
#N numero de hombres, se asume el número de mujeres es igual
def wPrefersM1OverM(prefer, w, m, m1, N): #indica si una mujer prefiere a M1 sobre M2, hombres
	for i in range(N): 
		if (prefer[w][i] == m1): 
			return True
		if (prefer[w][i] == m): 
			return False

def stableMarriage(prefer, N): #Implementa el algoritmo Gale-shapley asumiendo que los hombres se proponen a las mujeres 
  wPartner = [-1 for i in range(N)]
  mFree = [False for i in range(N)] 
  it = 0
  freeCount = N 
  while (freeCount > 0): 
    m = 0
    while (m < N): 
      if (mFree[m] == False): 
        break
      m += 1
    i = 0
    while i < N and mFree[m] == False: 
      it += 1
      w = prefer[m][i] 
      if (wPartner[w - N] == -1): 
        wPartner[w - N] = m 
        mFree[m] = True
        freeCount -= 1
      else: 
        m1 = wPartner[w - N] 
        if (wPrefersM1OverM(prefer, w, m, m1, N) == False): 
          wPartner[w - N] = m 
          mFree[m] = True
          mFree[m1] = False
        i += 1
  return it
```

```
#N = 4
i = 0
x = list(range(0,100))
while i < 100:
  n = [[]] * 8
  for j in range(0,8):
    n[j] = np.random.permutation(list(range(0,4)))
  x[i] = stableMarriage(n,4)
  i += 1
print("Distribución del número de iteraciones con n = 4: ")
plt.hist(x)
plt.xlabel("Iteraciones")
plt.ylabel("")
plt.show()
```
